```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

```python
X, y = load_iris(return_X_y=True)
```

```python
sse = [] #SUM OF SQUARED ERROR
for k in range(1,11):
    km = KMeans(n_clusters=k, random_state=2)
    km.fit(X)
    sse.append(km.inertia_)
```

```python
sns.set_style("whitegrid")
g=sns.lineplot(x=range(1,11), y=sse)

g.set(xlabel ="Number of cluster (k)", ylabel = "Sum Squared Error", title ='Elbow Method')

plt.show()
```

```python
kmeans = KMeans(n_clusters = 3, random_state = 2)
kmeans.fit(X)
```

```python
kmeans.cluster_centers_
```

```python
pred = kmeans.fit_predict(X)
pred
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.scatter(X[:,0],X[:,1],c = pred, cmap=cm.Accent)
plt.grid(True)
for center in kmeans.cluster_centers_:
    center = center[:2]
    plt.scatter(center[0],center[1],marker = '^',c = 'red')
plt.xlabel("petal length (cm)")
plt.ylabel("petal width (cm)")
    
plt.subplot(1,2,2)   
plt.scatter(X[:,2],X[:,3],c = pred, cmap=cm.Accent)
plt.grid(True)
for center in kmeans.cluster_centers_:
    center = center[2:4]
    plt.scatter(center[0],center[1],marker = '^',c = 'red')
plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
plt.show()
```

```python
from sklearn.cluster import MeanShift


```

```python
# Apply MeanShift clustering
meanshift = MeanShift()
meanshift.fit(X)
```

```python
# Get the cluster labels and centers
labels = meanshift.labels_
cluster_centers = meanshift.cluster_centers_
```

```python
df = pd.DataFrame(X, columns=['sepal length', 'sepal width', 'petal length', 'petal width'])
df['Cluster'] = labels
```

```python
# Visualize the clustering (on two features for simplicity)
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], s=100, c='red', label='Centers')
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.title('MeanShift Clustering on Iris Dataset')
plt.legend()
plt.show()
```

```python
# Visualize the clustering (on two features for simplicity)
plt.scatter(X[:, 2], X[:, 3], c=labels, cmap='viridis')
plt.scatter(cluster_centers[:, 2], cluster_centers[:, 3], s=100, c='red', label='Centers')
plt.xlabel('petal length' )
plt.ylabel('petal width')
plt.title('MeanShift Clustering on Iris Dataset')
plt.legend()
plt.show()
```
